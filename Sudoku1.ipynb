```python
for row in range(9):
    for column in range(9):
        if grid[row][column] == 0:
            zone = [grid[row][i] for i in range(9)]
            for count in range(len(possible[row][column])):
                if possible[row][column][count] in zone:
                    possible[row][column].remove(possible[row][column][count])
        
```
